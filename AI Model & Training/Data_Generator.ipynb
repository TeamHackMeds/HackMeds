```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("dataset.csv")
data.head(10)
```

```python
columns_to_drop = [f"Symptom_{i}" for i in range(7, 18)]
data.drop(columns=columns_to_drop, inplace=True)
```

```python
data.head(10)
```

```python
new_dataset = data
new_dataset.to_csv("new_dataset.csv")
```

```python
categorical_cols = data.select_dtypes(include=['object']).columns.tolist()
```

```python
num_synthetic = 20000
```

```python
synthetic_categorical = data[categorical_cols].apply(
    lambda x: x.sample(num_synthetic, replace=True).values
)
```

```python
data_synthetic = pd.DataFrame({**synthetic_categorical})
```

```python
new_synthetic_data = pd.concat([data, data_synthetic], ignore_index=True)
```

```python
new_synthetic_data.to_csv("new_synthetic_data.csv", index="False")
```
